# Análise - Dataset 3 (NCM/NCA Battery)
Carregando e explorando os dados baseados em tempo, tensão e corrente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

dataset_path = r'c:\Users\pedro\OneDrive\Documentos\pedro\projetps\AmazoniaSustentavel\data-driven-capacity-estimation-from-voltage-relaxation\Dataset_3_NCM_NCA_battery\Dataset_3_NCM_NCA_battery'
files = [f for f in os.listdir(dataset_path) if f.endswith('.csv')]
print(f'Total de arquivos CSV encontrados no dataset: {len(files)}')

if files:
    sample_file = os.path.join(dataset_path, files[0])
    print(f'Carregando {sample_file}...')
    df3 = pd.read_csv(sample_file)
    display(df3.head())
```

```python
if files:
    plt.figure(figsize=(10, 5))
    plt.plot(df3['time/s'], df3['Ecell/V'], label='Ecell/V')
    plt.title(f'Tensão da Célula vs Tempo - {files[0]}')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Tensão (V)')
    plt.grid(True)
    plt.legend()
    plt.show()
```

### Gráficos em Diferentes Intervalos de Tempo
Analisando o comportamento dinâmico da tensão em recortes de 5000 segundos.

```python
if files:
    t_max = df3['time/s'].max()
    interval_size = 5000
    intervalos = [(i, min(i + interval_size, t_max)) for i in range(0, int(t_max), interval_size)]
    
    print(f'Gerando {len(intervalos)} gráficos separadamente para evitar sobrecarga de memória...')
    for t_start, t_end in intervalos:
        mask = (df3['time/s'] >= t_start) & (df3['time/s'] <= t_end)
        df_subset = df3[mask]
        
        if not df_subset.empty:
            plt.figure(figsize=(10, 4))
            plt.plot(df_subset['time/s'], df_subset['Ecell/V'], color='orange')
            plt.title(f'Tempo: {t_start} a {t_end:.0f}s - {files[0]}')
            plt.xlabel('Tempo (s)')
            plt.ylabel('Tensão (V)')
            plt.grid(True)
            plt.tight_layout()
            plt.show()
```
