# Comparação de Colunas dos Datasets
Carregando uma amostra de cada dataset para comparar quais atributos estão presentes em comum e alertar sobre diferenças de formatação ou nomenclatura de colunas entre os datasets.

```python
import pandas as pd
import os

path1 = r'c:\Users\pedro\OneDrive\Documentos\pedro\projetps\AmazoniaSustentavel\data-driven-capacity-estimation-from-voltage-relaxation\Dataset_1_NCA_battery\Dataset_1_NCA_battery'
path2 = r'c:\Users\pedro\OneDrive\Documentos\pedro\projetps\AmazoniaSustentavel\data-driven-capacity-estimation-from-voltage-relaxation\Dataset_2_NCM_battery\Dataset_2_NCM_battery'
path3 = r'c:\Users\pedro\OneDrive\Documentos\pedro\projetps\AmazoniaSustentavel\data-driven-capacity-estimation-from-voltage-relaxation\Dataset_3_NCM_NCA_battery\Dataset_3_NCM_NCA_battery'

file1 = os.path.join(path1, [f for f in os.listdir(path1) if f.endswith('.csv')][0])
file2 = os.path.join(path2, [f for f in os.listdir(path2) if f.endswith('.csv')][0])
file3 = os.path.join(path3, [f for f in os.listdir(path3) if f.endswith('.csv')][0])

df1 = pd.read_csv(file1, nrows=1)
df2 = pd.read_csv(file2, nrows=1)
df3 = pd.read_csv(file3, nrows=1)

cols1 = set(df1.columns)
cols2 = set(df2.columns)
cols3 = set(df3.columns)

print('--- Dataset 1 (NCA) ---')
print(list(df1.columns))
print('\n--- Dataset 2 (NCM) ---')
print(list(df2.columns))
print('\n--- Dataset 3 (NCM/NCA) ---')
print(list(df3.columns))

print('\n--- Colunas Comuns aos Três Datasets ---')
print(list(cols1.intersection(cols2).intersection(cols3)))

print('\n--- Presentes apenas no Dataset 1 ---')
print(list(cols1 - cols2 - cols3))

print('\n--- Presentes apenas no Dataset 2 ---')
print(list(cols2 - cols1 - cols3))

print('\n--- Presentes apenas no Dataset 3 ---')
print(list(cols3 - cols1 - cols2))
```
